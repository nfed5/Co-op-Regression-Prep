# PostgresSQL + Python Connection

```python
import psycopg2 as pg2
```

# Step 1: Define Class and Methods

```python
class Database: 
    def __init__(self, dbname, user, password):
        self.dbname = dbname
        self.user = user
        self.password = password
    
    def connect(self):
        conn = pg2.connect(database = self.dbname, user = self.user, password = self.password)
        cur = conn.cursor()
        return cur
        
    def query_db(self, cur, query):
        cur.execute(query)
        results = [row for row in cur.fetchall()]
        cur.close()
        return results
```

# Step 2: Create Objects

```python
#create an object
dvdrental = Database('dvdrental', 'postgres', 'N1chola$')
```

```python
query = 'SELECT * FROM customer_info'
query2 = 'SELECT * FROM customer'
cur = dvdrental.connect()
returns = dvdrental.query_db(cur, query)
returns[1:11]
```
