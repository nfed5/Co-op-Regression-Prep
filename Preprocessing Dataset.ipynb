# Get_dummies

```python
#split the 'Reason for Absence' column into dummy variables
reasons = pd.get_dummies(data['Reason for Absence'])
pd.options.display.max_columns = None
reasons.head()
```

```python
#make sure each row only adds up to 1; 0 means missing value more than 1 not possible can only have one reason 
#for being absent 
reasons['check'] = reasons.sum(axis=1)
reasons.head()
```

```python
#we want to make sure there are only 1's in the 'check' column
#add up all the values in the 'check' column and you should get the number of rows
reasons['check'].sum(axis=0)
```

```python
#what if there are some 0's and 2's but we just are not seeing them?
#check to see that there are only 1's using .unique()
reasons['check'].unique()
```

```python
#now remove the 'check' column
dummy_set = reasons.drop(['check'], axis = 1)
dummy_set.head()
```

# Group the Dummies and Concatenate them

```python
#split the 'Reason for Absence' column into dummy variables
#drop the first column to avoid multicollinearity
reasons_dummie = pd.get_dummies(data['Reason for Absence'], drop_first=True)
pd.options.display.max_columns = None
reasons_dummies.head()
```

```python
reason_1 = reasons_dummie.loc[:, 1:14].max(axis=1)
reason_2 = reasons_dummie.loc[:, 15:17].max(axis=1)
reason_3 = reasons_dummie.loc[:, 18:21].max(axis=1)
reason_4 = reasons_dummie.loc[:, 22:].max(axis=1)
```

```python
new_dataset = pd.concat([data, reason_1, reason_2, reason_3, reason_4], axis = 1)
new_dataset.head()
```
